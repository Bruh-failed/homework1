Problem 10:

What is the average budget used for the top 10 grossing movies?


```python

conn = sqlite3.connect("movies.sqlite")

# What is this doing is that it is going to get the average budget of T where
# would be collected of the FROM stuff. Under the FROM, it will select the top
# 10 movies with revune and gets only 10 then gets the average using the T.budget


query = """SELECT AVG(T.budget) AS budget
FROM
  (SELECT revenue, budget
  FROM MOVIES
  ORDER BY revenue DESC
  LIMIT 10) as T
  """

df = pd.read_sql(query, conn)

df.head(10)
```
